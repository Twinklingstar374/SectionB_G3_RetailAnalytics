```python
import pandas as pd
df = pd.read_csv("new_retail_data.csv")
```

```python
shape = df.shape
shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.isna().sum()
```

```python
drop_cols = [
    'Name', 'Email', 'Phone', 'Address', 'Zipcode'
]
df.drop(columns=drop_cols, inplace=True)
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Total_Purchases'].fillna(0, inplace=True)
df['Amount'].fillna(df['Amount'].median(), inplace=True)
df['Total_Amount'].fillna(df['Total_Amount'].median(), inplace=True)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Customer_Segment'].fillna('Unknown', inplace=True)
df['Feedback'].fillna('No Feedback', inplace=True)
df['Order_Status'].fillna('Pending', inplace=True)
df['Payment_Method'].fillna('Unknown', inplace=True)
df['Shipping_Method'].fillna('Standard', inplace=True)
df['Income'].fillna(df['Income'].mode()[0], inplace=True)
```

```python
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Date'].fillna(method='ffill', inplace=True)

df['Time'].fillna('00:00:00', inplace=True)
```

```python
df['avg_order_value'] = df['Total_Amount'] / df['Total_Purchases']
df['avg_order_value'].fillna(0, inplace=True)
```

```python
def age_group(age):
    if age < 25:
        return 'Young'
    elif age < 45:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['Age'].apply(age_group)
```

```python
def purchase_type(x):
    if x <= 5:
        return 'Low'
    elif x <= 15:
        return 'Medium'
    else:
        return 'High'

df['purchase_category'] = df['Total_Purchases'].apply(purchase_type)
```

```python
def revenue_category(x):
    if x < 1000:
        return 'Low'
    elif x < 5000:
        return 'Medium'
    else:
        return 'High'

df['revenue_category'] = df['Total_Amount'].apply(revenue_category)
```

```python
df['Date'] = pd.to_datetime(df['Date'])

df['day'] = df['Date'].dt.day
df['weekday'] = df['Date'].dt.day_name()
df['is_weekend'] = df['weekday'].isin(['Saturday', 'Sunday'])
```

```python
df['order_success'] = df['Order_Status'].apply(lambda x: 1 if x == 'Completed' else 0)
```

```python
def rating_label(x):
    if x >= 4:
        return 'Good'
    elif x >= 2:
        return 'Average'
    else:
        return 'Poor'

df['rating_category'] = df['Ratings'].apply(rating_label)
```

```python
def customer_value(row):
    if row['Total_Amount'] > 5000 and row['Total_Purchases'] > 10:
        return 'High Value'
    elif row['Total_Amount'] > 2000:
        return 'Medium Value'
    else:
        return 'Low Value'

df['customer_value_segment'] = df.apply(customer_value, axis=1)
```
